```python
import requests
from bs4 import BeautifulSoup
from IPython.core.display import HTML
```

```python
year = 2015
type_code = "gs"
endpoint = "http://www.atpworldtour.com/en/scores/results-archive?year={}&tournamentType={}".format(year, type_code)
print endpoint
```

```python
req = requests.get(endpoint)
page = req.text
soup = BeautifulSoup(page, "lxml")
```

```python
tournaments_table = soup.find("table")
```

```python
rows = [row for row in tournaments_table.find_all("tr")]
print rows[0]
```

```python
[col for col in rows[1].find_all("td") if col.get_text()][-1].find("a").attrs["href"]
```

```python
HTML(unicode(rows[1]))
```

```python
tournaments = soup.find_all("ul", {'id' : "tournamentDropdown" })
assert len(tournaments)==1
tournaments = tournaments[0]
```

```python
for tournament in tournaments.find_all('li', {'data-value' : True}):
    code = tournament.attrs['data-value']
    name = unicode(tournament.string).strip().replace(' ', '-')
    print "http://www.atpworldtour.com/en/scores/archive/{}/{}/{}/results?matchType=singles".format(name, code, year)
    
```
